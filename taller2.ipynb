Un vendedor recibe un sueldo base más un 10% extra por
comisión de sus ventas, el vendedor desea saber cuánto
dinero obtendrá por concepto de comisiones por las tres
ventas que realiza en el mes y el total que recibirá en
el mes tomando en cuenta su sueldo base y comisiones.

```python
# sueldo base
sueldobase = int(input("digite el sueldo base: "))
# ventas
venta1 = int(input("digite el valor de la venta 1: "))
venta2 = int(input("digite el valor de la venta 2: "))
venta3 = int(input("digite el valor de la venta 3: "))
# calculo total ventas
ventas = venta1 + venta2 + venta3
print(f"el valor total de ventas = {venta1} + {venta2} + {venta3} es $ {ventas}")
# calculo de la comision 
comision = ventas * 0.10
print(f"la comision es $ {comision}")
# suma ventas + comision (total a pagar)
totalapagar = sueldobase + comision
print("sueldo base = ${:,.0f} ,total ventas = ${:,.0f} , comision = ${:,.0f}".format(sueldobase, ventas, comision))
print(f"el total a pagar es $ {totalapagar}")
```

Una tienda ofrece un descuento del 15% sobre el total
de la compra y un cliente desea saber cuánto deberá
pagar finalmente por su compra.

```python
# valor articulos
articulo1 = int(input("digite el valor del articulo 1: "))
articulo2 = int(input("digite el valor del articulo 2: "))
articulo3 = int(input("digite el valor del articulo 3: "))
articulo4 = int(input("digite el valor del articulo 4: "))
# total venta
totalcompra = articulo1 + articulo2 + articulo3 + articulo4
# descuento
descuento = totalcompra * 0.15
# total a pagar
totalapagar = totalcompra * 0.85
print("total compra = ${:,.0f} , descuento = ${:,.0f} , total a pagar = ${:,.0f}".format(totalcompra, descuento, totalapagar))



```

Un alumno desea saber cuál será su calificación final en
la materia de Algoritmos. Dicha calificación se compone
de tres exámenes parciales.

```python
# examenes parciales
parcial1 = float(input("digite el resultado parcial 1: "))
parcial2 = float(input("digite el resultado parcial 2: "))
parcial3 = float(input("digite el resultado parcial 3: "))
# calificacion final
calificacionfinal = round((parcial1 + parcial2+ parcial3) / 3, 2)

print(f"la calificacion final materia algortimos es: {calificacionfinal}")

```

Un maestro desea saber qué porcentaje de hombres y que
porcentaje de mujeres hay en un grupo de estudiantes.

```python
# cantidad de alumnos hombres y mujeres
hombres = int(input("digite la cantidad de hombres en el salon: "))
mujeres = int(input("digite la cantidad de mujeres en el salon: "))
# total de alumnos
totalalumnos = hombres + mujeres
# porcentaje de alumnos hombre / mujeres
porcentajehombres = hombres / totalalumnos
porcentajemujeres = mujeres / totalalumnos
print(f"el total de alumnos en el salon es: {totalalumnos}")
print("el porcentaje de hombres en el salon es:{} y el porcentaje de mujeres es: {}".format(porcentajehombres, porcentajemujeres))
```
